```python
tmp = pd.DataFrame()

tmp['is_week'] = list(map(int, [re.match('.*week.*',i) != None for i in data['age_upon_outcome']]))

tmp['is_month'] = list(map(int, [re.match('.*month.*',i) != None for i in data['age_upon_outcome']]))

tmp['is_year'] = list(map(int, [re.match('.*year.*',i) != None for i in data['age_upon_outcome']]))

tmp['nub_age'] = list(map(int, [re.findall('[0-9]+',i)[0] for i in data['age_upon_outcome']]))

tmp['age'] = (tmp['is_week'] * 7 + tmp['is_month'] * 30 + tmp['is_year'] * 365) * tmp['nub_age']

tmp.index = data.index
tmp
```

```python
data['age_days'] = tmp['age']
data.drop('age_upon_outcome', axis= 1,inplace = True)
data
```
